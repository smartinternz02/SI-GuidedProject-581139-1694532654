```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

# Data visualisation for car_crashes

```python
df=sns.load_dataset('car_crashes')
```

```python
df
```

```python
df.info()
```

```python
df.tail()
```

## Line Graph

```python
df.plot(x="alcohol",y="ins_losses")
```

## Grid


```python
df.plot(x="alcohol",y="ins_losses",ms=20)
plt.grid()
```

## Scatter Plot

```python
plt.scatter(x=df["alcohol"], y=df["ins_losses"])
plt.xlabel("Alcohol")
plt.ylabel("Insurance Losses")
plt.title("Scatter Plot of Alcohol vs. Insurance Losses")
plt.show()
```

```python
plt.scatter(x=df["alcohol"], y=df["ins_losses"],color='hotpink')
plt.xlabel("Alcohol")
plt.ylabel("Insurance Losses")
plt.title("Scatter Plot of Alcohol vs. Insurance Losses")
plt.show()
```

## Bar Chart

```python
x = np.array(df["alcohol"])
y = np.array(df["ins_losses"])
plt.xlabel("Alcohol")
plt.ylabel("Insurance Losses")
plt.title("Scatter Plot of Alcohol vs. Insurance Losses")
plt.bar(x,y)
```

```python
x = np.array(df["alcohol"])
y = np.array(df["ins_losses"])
plt.title("Scatter Plot of Alcohol vs. Insurance Losses")
plt.barh(x,y)
```

```python
x = np.array(df["alcohol"])
y = np.array(df["ins_losses"])
plt.title("Scatter Plot of Alcohol vs. Insurance Losses")
plt.barh(x,y,color='pink')
```

```python
x = np.array(df["alcohol"])
y = np.array(df["ins_losses"])
plt.title("Scatter Plot of Alcohol vs. Insurance Losses")
plt.bar(x,y,width=0.1)
```

## Histogram

```python
plt.hist(df['alcohol'])
plt.show
```

## Pie Chart

```python
import matplotlib.pyplot as plt

labels = ["Alcohol", "ins_losses"]
x = [60, 70]

fig, axes1 = plt.subplots(figsize=(10, 10))
axes1.pie(x, labels=labels, autopct='%0.2f%%', colors=["pink", "green"])
axes1.legend(labels, loc="best")

plt.show()
```
